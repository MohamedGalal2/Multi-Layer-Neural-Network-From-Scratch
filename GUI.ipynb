```python
import tkinter as tk
from tkinter import *
from tkinter import messagebox

root = Tk()
root.geometry("950x650")

L = Label(root, text="Iris Flower")
L.config(font =("Times 20 italic bold", 30))
L.pack()

sele_f = IntVar()
sele_c = IntVar()


feature_l = Label(root,text = "Select activation function:") 
feature_l.pack(anchor=W)
feature_l.config(font =("Courier", 14))


def activation_sel():
    
    selection = "You selected option " + str(sele_f.get())
    label.config(text = selection)

R1_f = Radiobutton(root, text="Sigmoid", variable=sele_f, value=1, command=activation_sel)
R1_f.pack( anchor = W )
R1_f.config(font =("Courier", 12))


R2_f = Radiobutton(root, text="Tanh", variable=sele_f, value=2, command=activation_sel)
R2_f.pack( anchor = W )
R2_f.config(font =("Courier", 12))



label = Label(root)
label.pack()


# Define Function to print the input value
def display_input():
    
    print("Bias: ", check_var.get())
    
    global hl
    hl= T_hl.get("1.0",END)
    print("Number of hidden layers: ", hl)
    
    global nn 
    nn= T_nn.get("1.0",END)
    print("Learning rate: ", nn)
    
    global lr 
    lr= T_lr.get("1.0",END)
    print("Learning rate: ", lr)
    
    global epochs
    epochs = T_epochs.get("1.0",END)
    print("Number of epochs: ", epochs)
    
    
    
    print("selected function: " + str(sele_f.get()))
    
    ff=sele_f.get()
    cc=sele_c.get()
    if( hl=="\n"):
        messagebox.showerror("Error", "Missing number of hidden layers")
    if( nn=="\n"):
        messagebox.showerror("Error", "Missing number of neurons")
    if(lr=="\n"):
        messagebox.showerror("Error", "Missing learning rate")
    if(epochs=="\n"):
        messagebox.showerror("Error", "Missing epochs")




# bias check box
check_var = IntVar() 
c = Checkbutton(root, text = "Bias",variable=check_var, onvalue=1, offvalue=0)
c.config(font =("Courier", 18))
c.pack()


# Create text widget and specify size.

#Create label for number of hidden layer
T_hl = tk.Text(root, bg = "light cyan" ,height = 2, width = 20) ##########

text_hl = Label(root, text = "Enter number of hidden layers")
text_hl.config(font =("Courier", 14))

text_hl.pack(anchor = W)
T_hl.pack(anchor = CENTER)

# Create label for number of neurons
T_nn = Text(root, bg = "light cyan" ,height = 2, width = 20)        ############

nn_l = Label(root, text = "Enter number of neurons")
nn_l.config(font =("Courier", 14))
nn_l.pack(anchor = W)

T_nn.pack(anchor = CENTER)

# Create label for learning rate
T_lr = tk.Text(root, bg = "light cyan" ,height = 2, width = 20) ##########

text_l = Label(root, text = "Enter learning rate")
text_l.config(font =("Courier", 14))

text_l.pack(anchor = W)
T_lr.pack(anchor = CENTER)


# Create label for epochs
T_epochs = Text(root, bg = "light cyan" ,height = 2, width = 20)        ############

epochs_l = Label(root, text = "Enter number of epochs")
epochs_l.config(font =("Courier", 14))
epochs_l.pack(anchor = W)

T_epochs.pack(anchor = CENTER)




# enter button
btn=tk.Button(root, text="Press Enter",background = "blue", fg = "white",
           command = display_input)
btn.pack(expand=True)
 

root.mainloop()
```
